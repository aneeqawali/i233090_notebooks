```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#dictionary making
paths = {
    "Path": ["W", "X", "Y", "Z"],
    "Distance": [300, 200, 150, 250],
    "Time": [25, 15, 20, 10],
    "Congestion": ["Medium", "Low", "Medium", "High"]
}

#dataframe making
df = pd.DataFrame(paths)

#congestion setting
congestion_mapping = {"Low": 1, "Medium": 2, "High": 3}
df['Congestion_Score'] = df['Congestion'].map(congestion_mapping)

print("Initial Environment State")
print(df)

#divide 100 to scale 
#time matters the most then congestion then distance
df['Utility'] = 100 / ( (df['Distance'] * 0.2) +  (df['Time'] * 0.5) + (df['Congestion_Score'] * 0.3) )

#max index sy max path 
best_path_index = df['Utility'].idxmax()
best_path = df.loc[best_path_index]

#print big to small
print("\nUtility Analysis")
print(df[['Path', 'Utility']].sort_values(by='Utility', ascending=False))

print(f"\nDecision: The Agent selects Path {best_path['Path']} with Utility {best_path['Utility']:.2f}")
```

```python
reality_df = pd.DataFrame({
    "Section": ["Section 1", "Section 2"],
    "Status": ["Healthy", "Wilted"] 
})
reality_df.set_index("Section", inplace=True)

model_df = pd.DataFrame({
    "Section": ["Section 1", "Section 2"],
    "Last_Observed_Status": ["Unknown", "Unknown"]
})
model_df.set_index("Section", inplace=True)

current_location = "Section 1"

print("Starting Model-Based Agent Simulation")

for step in range(1, 4):
    print(f"\n STEP {step} | Agent Location: {current_location}")
    observation = reality_df.loc[current_location, "Status"]
    
    model_df.loc[current_location, "Last_Observed_Status"] = observation
    print(f"Agent Update: I now remember {current_location} is {observation}")
    
    if model_df.loc[current_location, "Last_Observed_Status"] == "Wilted":
        action = "Water Plants"
        
        reality_df.loc[current_location, "Status"] = "Healthy"
       
        model_df.loc[current_location, "Last_Observed_Status"] = "Healthy"
        
    
    else:
        other_section = "Section 2" if current_location == "Section 1" else "Section 1"
        other_status = model_df.loc[other_section, "Last_Observed_Status"]
        
        if other_status == "Wilted":
            action = f"Move to {other_section}"
            current_location = other_section
        else:
            action = "Stay and Monitor"

    print(f"Action Taken: {action}")
    
    print("\nCurrent Internal Model")
    print(model_df)
    print("\nActual Greenhouse Reality")
    print(reality_df)
```
